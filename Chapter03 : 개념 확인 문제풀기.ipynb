<span style="font-size:80%">[문제2]<br><br></span>
```Python의 PEP 8은 무엇인지 설명하고 PEP 8의 핵심적인 규칙을 1개 이상을 설명해주세요```

<span style="color:red">답변</span>

PEP 8(Python Enhancement Proposal-8)은 Python의 공식 스타일 가이드로써 Python 커뮤니티에서 널리 따르는 규칙을 정리한 문서이다. Python 코드의 가독성을 높이고 일관된 코드 스타일을 유지하기 위해 개발되었다. 코드 작성 시 권장되는 포맷, 이름 규칙, 들여쓰기 방식 등을 명확하게 정의하고 있다. Python 코드의 표준을 제공하여 코드 리뷰나 협업 시 일관성을 유지하는 데 중요한 역할을 한다. 이러한 PEP 8의 핵심적인 규칙은 다음과 같다. 

첫째, 들여쓰기의 경우 네 개의 공백을 사용하며, 탭은 사용하지 않는 것을 권장한다. 

둘째, 변수와 함수의 이름은 소문자로 작성하며, 단어가 여러 개인 경우 밑줄(_)로 구분한다(snake_case). 

셋째, 클래스 이름의 경우 카멜 케이스(CamelCase)를 사용하며, 각 단어의 첫 글자를 대문자로 작성한다. 

넷째, 코드 라인의 길이는 79자를 넘지 않도록 하며 긴 라인은 여러 줄로 나누어 작성한다. 

다섯째, 연산자 양 옆에 공백을 두고, 함수 호출 시 인수 리스트에서 불필요한 공백을 피하여야 한다.

<span style="font-size:80%">[문제3]<br><br></span>
```Python에서 "#"이 의마하는 것은 무엇인지 설명해주세요```

<span style="color:red">답변</span>

Python에서 #은 주석(comment)를 의미한다. 주석은 코드에 대한 설명이나 메모를 작성할 때 사용되며 프로그램 실행에는 영향을 주지 않는다(Python 인터프리터에 의해 무시됨). 다만 디버깅 과정에서 일시적으로 코드의 시행을 막을 때 유용하게 사용될 수 있다. 코드의 가독성을 높이며, 다른 개발자들이 코드의 의도를 이해할 수 있도록 돕는 중요한 역할을 한다.

<span style="font-size:80%">[문제4]<br><br></span>
```Python의 들여쓰기 장점과 기능에 대해 설명해주세요```

들여쓰기(indentation)는 가독성을 향상한다. 들여쓰기를 통해 코드 구조가 직관적으로 구분되며, 코드의 흐름을 쉽게 이해할 수 있다. 또한 규칙적으로 강제되어, 코드 스타일이 일관되게 유지된다는 장점이 있다. 코드 블록의 범위를 명확히 정의하여, 코드 오류를 줄이는 데에도 도움이 된다. 

들여쓰기는 조건분(if), 반복문(for, while), 함수 등에 사용되어 코드 블록을 구분한다.

<span style="font-size:80%">[문제5]<br><br></span>
```print( )문의 기능 및 특징에 대해 설명해주세요```

<span style="color:red">답변</span>

print()는 데이터의 출력을 담당하는 함수로서, 전달된 값을 화면에 출력한다. 문자열, 숫자, 변수, 리스트 등 다양한 데이터를 출력할 수 있으며, 주로 프로그램의 실행 결과나 디버깅을 위해 값을 확인할 때 유용하다. 

print()는 다음과 같은 특징을 가진다. 첫째, 여러 값을 한 번에 출력할 수 있고 이때 출력되는 값들은 일반적으로 공백으로 구분된다. 둘째, 출력 후 자동으로 줄바꿈을 추가하는데, 이는 end 매개변수를 사용하면 변경할 수 있다. 셋째, sep 매개변수를 사용하여 출력되는 값 사이의 구분자를 변경할 수 있다. 넷째, f-string, .format() 등의 포맷팅되어 가독성 있는 출력도 가능하다.
